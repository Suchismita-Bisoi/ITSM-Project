# Model Creation

## 1. RandomForest 

```python
model = RandomForestClassifier(n_estimators= 250 , random_state = 10 , criterion = 'gini')
model.fit(X_train , y_train)
print(model.feature_importances_)
pd.DataFrame(model.feature_importances_,index = X.columns).sort_values(0,ascending=False)
```

```python
y_train_pred = model.predict(X_train)

y_pred = model.predict(X_test)

print(accuracy_score(y_pred,y_test)*100)
```

```python
print("\n Training accuracy = " ,accuracy_score(y_train,y_train_pred)*100)

print("\n Testing accuracy = " ,accuracy_score(y_test,y_pred)*100)
```

```python
print("Precision Score : {} ".format(precision_score(y_test,y_pred, average="weighted")*100))
```

```python
print("Recall Score : {} ".format(recall_score(y_test,y_pred, average="weighted")*100))
```

```python
print("F1 Score : {} ".format(f1_score(y_test,y_pred, average="weighted")*100))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
pd.crosstab(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

* From randomforest model we get the training accuracy of 100% and testing accuracy upto 86%.The model perform very well
